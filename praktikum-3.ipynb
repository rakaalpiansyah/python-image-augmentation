Bagian 1: Kode Contoh Modul

```python
import cv2
import matplotlib.pyplot as plt
image = cv2.imread('gambar1.jpg')

if image is None:
    print("Error: Gambar 'gambar1.jpg' tidak ditemukan. Pastikan nama dan lokasi file benar.")
else:
    resized = cv2.resize(image, (300, 300))

    cropped = image[50:250, 50:250]
    
    flipped = cv2.flip(image, 1)

    (h, w) = image.shape[:2]
    center = (w / 2, h / 2)
    M = cv2.getRotationMatrix2D(center, 270, 0.5)
    rotated = cv2.warpAffine(image, M, (w, h))

    plt.figure(figsize=(15, 5)) 

    plt.subplot(1, 4, 1)
    plt.imshow(cv2.cvtColor(resized, cv2.COLOR_BGR2RGB))
    plt.title('Resized')
    plt.axis('off')

    plt.subplot(1, 4, 2)
    plt.imshow(cv2.cvtColor(cropped, cv2.COLOR_BGR2RGB))
    plt.title('Cropped')
    plt.axis('off')

    plt.subplot(1, 4, 3)
    plt.imshow(cv2.cvtColor(flipped, cv2.COLOR_BGR2RGB))
    plt.title('Flipped')
    plt.axis('off')

    plt.subplot(1, 4, 4)
    plt.imshow(cv2.cvtColor(rotated, cv2.COLOR_BGR2RGB))
    plt.title('Rotated')
    plt.axis('off')

    plt.tight_layout()
    plt.show()
```

Bagian 2: Penyelesaian Tugas Mandiri

```python
import cv2
import os
import matplotlib.pyplot as plt

daftar_gambar = [
    'gambar1.jpg', 'gambar2.png', 'gambar3.jpg', 'gambar4.jpg', 'gambar5.jpg',
    'gambar6.jpg', 'gambar7.jpg', 'gambar8.jpg', 'gambar9.jpg', 'gambar10.jpg'
]

folder_output = 'hasil_augmentasi'
if not os.path.exists(folder_output):
    os.makedirs(folder_output)
    print(f"Folder '{folder_output}' berhasil dibuat.")

print("Memulai proses augmentasi 10 gambar...\n")

for nama_file in daftar_gambar:
   
    img = cv2.imread(nama_file)
    
    if img is None:
        print(f"[Lewati] {nama_file} tidak ditemukan di direktori.")
        continue
    
    img_resized = cv2.resize(img, (400, 400))
    img_cropped = img_resized[100:300, 100:300]
    img_flipped = cv2.flip(img, 0)
    
    (h, w) = img.shape[:2]
    center = (w // 2, h // 2)
    M = cv2.getRotationMatrix2D(center, 45, 1.0)
    img_rotated = cv2.warpAffine(img, M, (w, h))
    
    nama_dasar, ekstensi_asli = os.path.splitext(nama_file)
   
    cv2.imwrite(f"{folder_output}/resized_{nama_dasar}.jpg", img_resized)
    cv2.imwrite(f"{folder_output}/cropped_{nama_dasar}.jpg", img_cropped)
    cv2.imwrite(f"{folder_output}/flipped_{nama_dasar}.jpg", img_flipped)
    cv2.imwrite(f"{folder_output}/rotated_{nama_dasar}.jpg", img_rotated)
    
    print(f"[Sukses] 4 variasi manipulasi untuk '{nama_file}' telah disimpan dalam format .jpg.")

print("\nSeluruh proses selesai! Silakan periksa folder 'hasil_augmentasi'.")
```
